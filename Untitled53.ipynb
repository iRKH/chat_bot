```python
import pandas as pd
import datetime as dt
```

```python
import warnings
warnings.filterwarnings("ignore")
```

```python
# df = pd.read_csv('/DataSets/LP 02.txt', parse_dates=['timestamp'], sep=',')
# df = pd.read_csv('/DataSets/LP 02.txt', sep=',')

url = 'https://drive.google.com/file/d/1YQrrCEwdNEE5YvuBpaXTAJ5q3F1B5jXL/view?usp=sharing'
path = 'https://drive.google.com/uc?export=download&id='+url.split('/')[-2]
df = pd.read_csv(path, sep=',')
df[:2]
```

```python
df.info()
```

```python
df = (
        df
        .rename(columns={'clientId':'client_id', 'log':'log_duration'})
        .astype({'client_id':'int'})
     )
df
```

```python
df = df[df['client_id'] != 1].reset_index(drop=True)
df
```

```python
# при загрузке можно использовать параметр parse_dates=['timestamp']

df['timestamp'] = pd.to_datetime(df['timestamp'], infer_datetime_format=True).dt.round('1s')
df['timestamp'] = df['timestamp'].apply(lambda x: dt.datetime.strftime(x, "%d-%m-%Y %H:%M:%S"))
df['timestamp'][:2]
```

```python
df['log_duration'] = df['log_duration'].str.split(':').str[1].str.split(' ').str[0].astype({'log_duration':'int'})
df
```

```python
df = df[(df['log_duration'] > 15)].reset_index(drop=True)
df
```

```python
df = df.query('15 < log_duration < 59').reset_index(drop=True)
df
```
